## Merge

```python
df_merge1 = pd.read_csv("3-merge_data1.csv")
```

```python
df_merge1
```

```python
df_merge2 = pd.read_csv("3-merge_data2.csv")
```

```python
df_merge2
```

```python
# merge - inner join
df_merged_inner = pd.merge(
    left = df_merge1,
    right= df_merge2,
    on = 'Employee_ID',    # Hangileri üzerinden eşleşme sağlayıp birleştireceğimizi seçiyor
    how = 'inner'
)
```

```python
df_merged_inner
```

```python
# merge - outer join
df_merged_outer = pd.merge(
    left = df_merge1,
    right = df_merge2,
    on = 'Employee_ID',
    how = 'outer'
)

df_merged_outer
```

```python
# merge - left join
df_merged_left = pd.merge(
    left = df_merge1,
    right = df_merge2,
    on = 'Employee_ID',
    how = 'left'
)

df_merged_left
```

```python
# merge - right join
df_merged_right = pd.merge(
    left = df_merge1,
    right = df_merge2,
    on = 'Employee_ID',
    how = 'right'
)

df_merged_right
```

## Apply

```python
df = pd.read_csv("3-apply_function_data.csv")
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
# Örnek 1: Maaşlarına göre işçileri sınıflandıran bir fonksiyon 
def salary_category(salary):
    if salary < 50000:
        return 'Low'
    elif 5000 <= salary < 80000:
        return 'Medium'
    else:
        return 'High'
```

```python
df['Salary_Category'] = df['Salary'].apply(salary_category)
```

```python
df
```

```python
# Örnek 2: Tecrübelerine göre işçilerin performans skorlarını 1 arttıran fonksiyon
def adjust_performance(df):
    if df['Experience'] > 10:
        df['Performance_Score'] += 1
    return df['Performance_Score']
```

```python
df['Adjusted_Performance'] = df.apply(adjust_performance, axis = 1)

df
```

```python
# Örnek 3: İsim kısmında _ yerlerine boşluk koyan bir uygulama

df['Formatted_Names'] = df['Name'].apply(lambda x: x.replace('_', ' '))
```

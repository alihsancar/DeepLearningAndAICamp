# Pandas Series'e Giriş

```python
import numpy as np
import pandas as pd
```

```python
grade = { "Ali" : 100, "Melih" : 90, "Feza" : 70 }
```

```python
pd.Series(grade)
```

```python
names = ["Ali", "Melih", "Feza"]
grades = [100, 90, 70]
```

```python
pd.Series(names)
```

```python
pd.Series(grades)
```

```python
pd.Series(data = grades, index = names)
```

## Numpy ile Pandas Serileri

```python
numpy_array = np.array([50,40,30,20])
```

```python
pd.Series(numpy_array)
```

## Pandas Serieslerde Aritmetik İşlemler

```python
result_1 = pd.Series(data= [100, 90, 70], index=["Ali", "Melih", "Feza"])
result_2 = pd.Series(data= [80, 80, 70], index=["Ali", "Melih", "Feza"])
```

```python
result_1["Ali"]
```

```python
result_2["Melih"]
```

```python
result_1 + result_2
```

```python
result_1 * result_2
```

```python
result_1 - result_2
```

```python
result_1 / result_2
```

### İndexleri Farklı Olan Pandas Seriesleri Toplama

```python
different_series_1 = pd.Series([20,30,40,50],["a","b","c","d"])
different_series_2 = pd.Series([20,10,40,70],["a","c","f","g"])
```

```python
different_series_1 + different_series_2
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("5-athlete_events.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull()
```

```python
data.columns
```

```python
data.info()
```

```python
plt.scatter('Height', 'Weight', data = data)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Weight vs Height')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(x = 'Height', y = 'Weight', data = data)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Weight vs Height')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x = 'Height', y = 'Weight', hue = 'Sex', data = data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = 'Height', y = 'Weight', hue = 'Sex', style = 'Medal', data = data)
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x = 'Height', y = 'Weight', hue = 'Sex', style = 'Medal', size = 'Age', data = data)
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.lineplot(x = 'Height', y = 'Weight', hue = 'Sex', data = data)
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.displot(x = 'Height', hue = 'Sex', data = data)
plt.ylabel('Frequency')
plt.title('Atheletes Height Histogram')
plt.show()
```

```python
sns.set_style('white')
sns.displot(x = 'Height', hue = 'Sex', data = data, kind = 'kde')
plt.ylabel('Frequency')
plt.title('Atheletes Height Histogram')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x='Height', y='Weight', hue="Sex",data=data, kind="kde")
plt.ylabel('Frequency')
plt.title('Atheletes Height Histogram')
plt.show()
```

```python
sns.barplot(x='Medal', y='Height', hue="Sex", data=data)
plt.title('Medals by Height')
plt.show()
```

```python
sns.catplot(x='Medal', y='Height', hue='Sex', col='Season', data=data, kind='bar')
```

```python
data.corr(numeric_only=True)
```

```python
sns.heatmap(data.corr(numeric_only=True), annot = True)
```

```python
data = np.array([5, 7, 9, 15, 20, 22, 25, 30, 32, 35, 37, 40, 50, 55, 60, 100])

plt.figure(figsize = (6,4))
sns.boxplot(data)
plt.title("Box Plot of Simple Numeric Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
df = sns.load_dataset("titanic")
df
```

```python
plt.figure(figsize = (15,5))

# Yaş'a göre boxplot
plt.subplot(1, 2, 1)
sns.boxplot(x = 'class', y = 'age', data = df)
plt.title("Box Plot of Age by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.grid(True)

# Ödenen Ücrete göre box plot
plt.subplot(1, 2, 2)
sns.boxplot(x = 'class', y = 'fare', data = df)
plt.title("Box Plot of Fare by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.yscale("log")  # Log scale for better visualization of high fares
plt.grid(True)

plt.show()
```

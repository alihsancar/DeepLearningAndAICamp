```python
data = np.array([10, 12, 13, 15, 18, 25, 100, 105])
```

```python
data.mean()
```

```python
data.std()
```

```python
z_scores = (data - data.mean()) / data.std()
```

```python
outliers = data[np.abs(z_scores > 1)]
outliers
```

2x + 3y = 8 and 5x + 7y = 19

```python
# coefficents
A = np.array([[2,3], [5,7]])

# constants
b = np.array([8, 19])
```

```python
solution = np.linalg.solve(A, b)
```

```python
solution
```
